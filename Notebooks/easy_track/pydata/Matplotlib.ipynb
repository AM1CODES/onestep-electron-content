# Plotting our first graph
---
Lets start off with a very simple example. Our goal is to plot a graph to show relation between numbers and their squares. We will make use of **'np.linspace()'** to make our job  easier.

* **Commands -** 

1. np.linspace(X,Y) \[NumPy] - It returns evenly spaced number over a specified interval. We need to manually set the start(X) and the end point(Y).
2. plt.plot(X,Y) \[Matplotlib] - It helps us plot the graph. We will plot the graph of x v/s y where x conatains the given set of numbers and y is its squares.
3. plt.show() \[Matplotlib] - It helps us to print/show ouurr graph when we run the cell.

These  are the commands that we will be using, to make our very first plot.

```python
X = np.linspace(0,5)
Y = X**2
plt.plot(X,Y)
plt.show()
```

And just like that we were able to plot our first graph but it looks very incomplete isn't it? So lets make it more informative by adding the axis labels for the x-axis and the y-axis and lets also give this a title to make it easier for the reader to understand our graph.

# Adding axis details and titles to the plots
---
* **Commands** - 
* plt.xlabel('x-axis') - This will give your x-axis a label of your choice.
* plt.ylabel('y-axis') - This will give your y-axis a label of your choice.
* plt.title('title') - This will give your  whole graph the title of your choice.

```python
X = np.linspace(0,5)
Y = X**2
plt.plot(X,Y)
plt.xlabel('Numbers(X)')
plt.ylabel('Square of Numbers(Y)')
plt.title('Number v/s Square')
plt.show()
```

And with that addition our plot is much more readable for anyone who is takig a look at it for the first time. We will now learn some more tricks that will help in making better plots for our data.

# Creating Figure and Axes
---
In this part we will see how we can create figures and axes. Figure can be imaginedd as the whole graph, with subplots, axes, legends, labels,etc. On the other hand Axes is just the area in which your plot appears.

* **Commands** - 
* plt.figure() - Helps in creating a canvas upon which we make our plots.
* .add_axes(\[left,bottom width,height]) - Helps us in adding an axes which has the following parameters in it.
* axes.set_xlabel('x-axis') - This will give your x-axis a label of your choice.
* axes.set_ylabel('y-axis') - This will give your y-axis a label of your choice.
* axes.set_title('title') - This will give your  whole graph the title of your choice.

Lets go ahead and apply these.

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
X = np.linspace(0,5)
Y = X**2
axes.plot(X,Y)
axes.set_xlabel('Numbers(X)')
axes.set_ylabel('Square of Numbers(Y)')
axes.set_title('Number v/s Square')
plt.show()
```

# Inserting Multiple Axes
---
We now know how to create an axes. Lets create more axes and actually play with the different parameters i.e. left,bottom,width,height

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])
X = np.linspace(0,5)
Y = X**2
axes1.plot(X,Y)
axes2.plot(Y,X)
axes1.set_xlabel('Numbers(X)')
axes2.set_xlabel('Numbers(X)')
axes1.set_ylabel('Square of Numbers(Y)')
axes2.set_ylabel('Square of Numbers(Y)')
axes1.set_title('Large plot')
axes2.set_title('Small plot')
plt.show()
```

And just like that we were able to create two plots in one figure by adjusting the list of parameters that we pass in the axes. We strongly recommend you to try and change these parameters and see what you can come up with.

# Creating Subplots
---
We can divide our plots into different numbers by creating subplots. It helps us accomodate more number of plots in a confined canvas.

* **Commands**-
* plt.subplots(nrows,ncols) - We will use this to create the subplots we need

```python
fig = plt.figure()
fig,axes = plt.subplots(nrows=1,ncols=2)
X = np.linspace(0,5)
Y = X**2
for current_ax in axes:
    ax = current_ax.plot(X,Y)
plt.show(ax)
```

And that's how you create multiple plots in one figure/canvas.
*Note* - If you see that your axes are overlapig or you are not happy with the spacing of the plots,
use **'plt.tight_layout()'** to prevent overlaping of the axes.

# Chaging the Figure size
---
Sometimes there's a lot of data that you wish to visualize but you are falling short of canvas or figure space to accomodate it all, here we can give **'figsize'** as a parameter in 'plt.figure()' to set the size of the graph we wish to plot.

* **Command** - 
* plt.figure(figsize = (a,b)) = We can set the size of the graph we wish to have using this command. Here, a is the width and b is the height.

```python
X = np.linspace(0,5)
Y = X**2
fig = plt.figure(figsize=(5,2))
ax = fig.add_axes([0,0,1,1])
ax1 = ax.plot(X,Y)
ax.set_xlabel('Numbers(X)')
ax.set_ylabel('Square of Numbers(Y)')
ax.set_title('Numbers v/s Squares')
plt.show(ax1)
```

# Use of Legends in Graphs
---
In your graphs, if you dont describe what the lines that you just plotted depict, then it wouuld get really tough for the reader to understand what exactly a line describes. Legends make it easier for the reader to figure out what the lines in the plots actually describe.

* **Command** - 
* **ax.legend()/plt.legend()** = This will help in creating a legend for your graph and makes it easier to read. You can also pass in a parameter **'loc'** and set it to best and it will find the best place for your legend to be placed in the entire figure.

```python
X = np.linspace(0,5)
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(X,X**2,label = 'X-Squared')
ax.plot(X,X**3,label = 'X-Cubed')
ax.legend(loc='best')
plt.show()
```

# Changing the colour of the line and its width and style
---
In this we will see how we can add a few more parameters to change the width of the line, the colour of the line and its style.
**The parameters that we need to add are:-**
* color - Helps in changing color of the line.
* linewidth - Changes the width of the line.
* linestyle -  Changes the style of the line for eg. dashed lines.


```python
X = np.linspace(0,5) 
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(X,X**2,label = 'X-Squared',color = 'red',linewidth = 5)
ax.plot(X,X**3,label = 'X-Cubed',color = 'purple',linewidth = 2,linestyle = '--')
ax.legend()
plt.show()
```

In the above graph we applied all three properties and we saw our result. Now lets also add markers to see the data points clearly

# All about Markers
---
In this section we will learn about how we can add markers in our plot and to make the data points much more significant and we will also take a look at some of the properties that we can add to these markers to make it look much more significant.

* **Commands** - 
* marker - This will help you put the marker on the data point.
* markersize - This will help you  increase or decrease the size of the marker.
* markerfacecolor - Lets you change the face color of the marker.
* markeredgewidth  - Lets you change the edge width of the marker.
* markeredgecolor - Lets you  change the color of the edge of the marker

```python
X = np.linspace(0,5)
Y = X**2
fig = plt.figure(figsize = (10,5))
ax = fig.add_axes([0,0,1,1])
ax.plot(X,Y,color = 'purple',linewidth = 5,marker='o',markersize = 10,markerfacecolor = 'yellow',markeredgewidth = 3,markeredgecolor = 'Black')
ax.set_xlabel('Numbers(X)')
ax.set_ylabel('Square of Numbers(Y)')
ax.set_title('Numbers v/s Squares')
plt.show(ax)
```

# Importing Seaborn
---
We will start of by importing our library.
* **Command** -
* import seaborn as sns - This will import the library for us


```python
import seaborn as sns
```

# Loading the Dataset
---
To Visualize something, we first need the data and we can either use Pandas to import our data set or just use Seaborn to get the data. Seaborn has a few data sets that you canload by just calling the 'load_dataset' method.
* **Command** - 
* sns.load_dataset() - This lets you load the data set of your choice. You can pass the name of the data set that you need inorder to get the data.

```python
tips = sns.load_dataset('tips')
tips.head()
```

# Distplot
---
The very first plot that we will see is the 'distplot'. It basically creates a histogram and depicts overall distribution of continuous data variables. It also has a parameter called 'kde' or Kernel Density Estimation which is a way to estimate probability density function of a continuous random variable. So lets go ahead and apply these.

* **Command** - 
* sns.distplot() - Helps in creating the distribution plot with the kde line. We need to pass the column or the data in it to be plotted.

```python
sns.set(style="whitegrid") #helps in setting the grid background for the plots
sns.distplot(tips['total_bill'])
```

We can remove the kde line if we want by assigning the kde = False.

```python
sns.distplot(tips['total_bill'],kde=False)
```

# Joint Plot
---
We can consider it to be a combination of both scatter plot and histogram. It helps us in comparing two vaariables. Lets see how we can create it.

* **Command** - 
* sns.jointplot(x,y,data,kind) - Helps us in making the joint plot. We can pass the variable that we want on the x-axis in x and similarly for y. Then we pass our variable/object that has our data in it. We can give the parameter of kind if we want to change the kind of plot we need

```python
sns.jointplot(x='total_bill',y = 'tip',data = tips) #without kind
```

```python
sns.jointplot(x='total_bill',y = 'tip',data = tips,kind = 'hex') #with kind = hex
```

```python
sns.jointplot(x='total_bill',y = 'tip',data = tips,kind = 'reg') #with kind = reg
```

And this how you can customise your visualizations using the kind parameter.

# Bar Plot
---
Its a categorical plot.  It shows relation between categorical data and atleast one numerical column. A bar plot represents an estimate of central tendency for a numeric variable with the height of each rectangle and provides some indication of the uncertainty around that estimate using error bars.

* **Command** - 
* sns.barplot(x,y,data) -  Helps us in making the Bar plot. We can pass the variable that we want on the x-axis in x and similarly for y. Then we pass our variable/object that has our data in it.

```python
sns.barplot(x='sex',y='total_bill',data = tips)
```

We can also rotate these bars in order to have a different look at it. This helps a lot when you have too many bars. It just makes it easier for you to analyse the data.

```python
sns.barplot(x='total_bill',y='sex',data = tips)
```

# Count plot
---
Similar to barplot, we also have *'countplot()'*.It basically shows the counts of observations in each categorical bin using bars. It's basically a histogram over categorical data instead of quantitative data. Here's how you can apply it.

* **Command** - 
* sns.countplot(x,data) = Helps us in creating the countplot. We just need to give the input for variable x and pass the variable/object that has our data. The estimator will automatially count the number of different observations in our x.

```python
sns.countplot(x='sex',data = tips) #counting number of males and females using 'sex' column
```

# Heatmap
---
It's basically a colorful representation of data values present in the data set. It can also be used to find correlation between different columns. Lets check out it's implementation - 

* **Command** - 
* sns.heatmap() - This will help us in making our Heatmap. We just need to pass in our data and thats it! We will use this to get the correlation between the various columns in our data. We can use *'annot = True'* if we wish to see the data values in the heatmap as well.

```python
sns.heatmap(tips.corr(),annot = True)
```

# Box Plot
---
These are very basic plots used to visualize distributions. It is useful in comapring data between two groups. In Boxplot, the box shows the quartiles of the data and the extensions(whisker) show the rest of the distribution. The plots contain the minimum, First quartile[lower quartile] ,Second quartile[Median],Third quartile[Upper quartile] and maximum. It also helps you in getting the outliers in your data. Lets go ahead and apply it - 

* **Command** - 
* sns.boxplot(x,y,data) - Helps us in making the Box plot. We can pass the variable that we want on the x-axis in x and similarly for y. Then we pass our variable/object that has our data in it.

```python
sns.boxplot(x="day", y="total_bill", data=tips) #comapring total bill when compared to the day of the week
```

Plotting this

```python
sns.boxplot(x="total_bill", y="day", data=tips)
```

# Line Plot
---
This is one of the most common plots you will come across. It is basically used to show the relation between two variabkes x and y for various subsets. It depicts the relationship between continuous as well as categorical values in a continuous data point format.

* **Command** - 
* sns.lineplot(x,y,hue,data) - Helps us in making line plot.We can pass the variable that we want on the x-axis in x and similarly for y. Then we pass our variable/object that has our data in it. Hue adds and extra susbet of the columns for comaprison. We will compare and see the total_bill paid v/s tip and compare  it between males and females.

```python
sns.lineplot(x='total_bill',y='tip',hue = 'sex',data = tips)
```

# Violin Plot
---
Its one of the rare plots and combines the features of Box plot and kde plots. It is used to visualize distribution of numerical values of different variables. It is very similar to box plot but it has more emphasis on density. It is usually used when the amount of data is huge and showing individual observations is not possible.

* **Command** - 
* sns.violinplot(x,y,data) - Helps us in creating our violin plot. We can pass the variable that we want on the x-axis in x and similarly for y. Then we pass our variable/object that has our data in it.

```python
sns.violinplot(x="day", y="total_bill", data=tips)
```

We can also make use of 'Hue' that  we saw earlier to compare based on some other feature as well.When using hue nesting with a variable that takes two levels, setting 'split = True' will draw half of a violin for each level. This can make it easier to directly compare the distributions.

```python
sns.violinplot(x="day", y="total_bill", hue="smoker",
                    data=tips,split = True)
```

# Scatter Plot
---
It is one of the most common plots that you will come across. It compares two variables by plotting the data points. We can also compare other features by using the 'hue' and it really helps in seeing how your data is distributed.

* **Command** - 
* sns.scatterplot(x,y,data,hue) - Helps us in making the scatter plot.We can pass the variable that we want on the x-axis in x and similarly for y. Then we pass our variable/object that has our data in it. We can use the 'hue' to divide these scattered points into further categories.

```python
sns.scatterplot(x='total_bill',y='tip',data = tips,hue = 'time')
```

# Pair Plot
---
The last plot that we will be taking a look at is pair plot. It basically compares diffrent columns with each other by creating histograms, scatterplot, kde plot,etc. Lets take a look at how we can make it. It will compare all the columns with numerical data for you.

*Note* - *Your plot might take some time to  load depending on the machine you are using so wait for a few seconds before taking any actions.*

* **Command** - 
* sns.pairtplot(data) - Helps us in making our pairplot. We just need to pass in our data to make this plot. We can also make use of 'hue' if we wana do the comparisons on other basis as well. 

```python
sns.pairplot(tips)
```

```python
sns.pairplot(tips,hue = 'sex') #adding a comparison feature of sex in the pairplot
```
